```python
import numpy as np
import pandas as pd
import scipy.io as sio
```

```python
data = sio.loadmat('./ex6data1.mat')
X, y = data['X'], data['y']
y = y.reshape(y.shape[0], )
X.shape, y.shape
```

# Visualize data

```python
import matplotlib.pyplot as plt
plt.subplots(figsize=(12,9))
plt.scatter(X[:, 0], X[:, 1], c=y)
```

# Split training and testing sets

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

# Try differen C
## hinge loss
The hinge loss function is the $cost_{0}(\theta^{T}x^{(i)})$ and $cost_{1}(\theta^{T}x^{(i)})$ functions stated in the video.
$$
l(y) = max(0, 1-t \cdot y)
$$

```python
from sklearn.svm import LinearSVC

svc = LinearSVC(C=1000, loss='hinge', max_iter=200000)
svc.fit(X, y)
svc.score(X, y)
```

```python
from mlxtend.plotting import plot_decision_regions
plt.subplots(figsize=(12,9))
ax = plot_decision_regions(X=X, y=y, clf=svc)
ax.set_xlim([X[:,0].min()-0.05, X[:,0].max()+0.05])
ax.set_ylim([X[:,1].min()-0.05, X[:,1].max()+0.05])
plt.title('SVM Decision Region Boundary', size=6)
plt.show()
```

```python
svc = LinearSVC(C=1, loss='hinge', max_iter=200000)
svc.fit(X, y)
svc.score(X, y)
```

```python
plt.subplots(figsize=(12,9))
ax = plot_decision_regions(X=X, y=y, clf=svc)
ax.set_xlim([X[:,0].min()-0.05, X[:,0].max()+0.05])
ax.set_ylim([X[:,1].min()-0.05, X[:,1].max()+0.05])
plt.title('SVM Decision Region Boundary', size=6)
plt.show()
```
