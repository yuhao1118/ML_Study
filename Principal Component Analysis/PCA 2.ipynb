```python
import numpy as np
import scipy.io as sio
```

# Import dataset

```python
data = sio.loadmat("./ex7faces.mat")
X = data["X"]
X.shape
```

# Plot 100 samples

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,8))

for index, (image,) in enumerate(zip(X[:100])):
    plt.subplot(10, 10, index + 1)
    plt.imshow(image.reshape(32,32).T, cmap=plt.cm.gray)
    plt.xticks(np.array([]))
    plt.yticks(np.array([]))
```

# Import PCA implementation from sklearn
- n_components: Number of components to keep i.e the dimension we want to reduce to

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=16)
Z = pca.fit_transform(X)
Z.shape
```

```python
X_recovered = pca.inverse_transform(Z)
X_recovered.shape
```

```python
plt.figure(figsize=(8,8))

for index, (image,) in enumerate(zip(X_recovered[:100])):
    plt.subplot(10, 10, index + 1)
    plt.imshow(image.reshape(32,32).T, cmap=plt.cm.gray)
    plt.xticks(np.array([]))
    plt.yticks(np.array([]))
```
