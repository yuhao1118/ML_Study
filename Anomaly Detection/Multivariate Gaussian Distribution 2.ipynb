# Multivariate Gaussian Distribution
The EllipticEnvelope of Sklearn implementation assumes features statify gaussian distribution.

```python
import numpy as np
import scipy.io as sio
```

# Import dataset

```python
data = sio.loadmat("./ex8data1")
X = data["X"]
CV = data["Xval"]
y_CV = data["yval"]
X.shape, CV.shape, y_CV.shape
```

# Visualizing data

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))
plt.scatter(X[:, 0], X[:, 1])
plt.xlabel("Latency")
plt.ylabel("Throughput")
```

# Using Sklearn to detect outliers
- contamination: The amount of contamination of the data set, i.e. the proportion of outliers in the data set.

```python
from sklearn.covariance import EllipticEnvelope
ctt_frac = np.sum(y_CV == 1) / y_CV.shape[0]
clf = EllipticEnvelope(contamination = ctt_frac)
y_pred = clf.fit_predict(X)
y_pred.shape
```

# Evaluating score
EllipticEnvelope treat 1 as normal data and -1 as outliers whereas we treat 0 as normal data and 1 as outliers.

```python
y_CV_fix = np.where(y_CV == 0, 1, -1).reshape(y_CV_fix.shape[0],)
clf.score(CV, y_CV_fix)
```

# Choosing classifier
As the comtamination proportion affects the threshold, we can follow the same idea of how we choose threshold $\epsilon$ in our own implementation to choose the best classifier.

```python
def select_clf(X, y_CV_fix, max_iters = 100):
    besf_clf = None
    best_score = 0
    step = 0.5 / max_iters
    for i in np.arange(0, 0.5, step):
        clf = EllipticEnvelope(contamination = i)
        y_pred = clf.fit(X)
        score = clf.score(CV, y_CV_fix)
        
        if score > best_score:
            best_score = score
            best_clf = clf
            
    return best_clf, best_score
```

```python
clf, score = select_clf(X, y_CV_fix)
y_pred = clf.predict(X)
score
```

```python
X_outliers = np.where(y_pred == -1) # indexs of the outliers
plt.figure(figsize=(10,8))
plt.scatter(X[:, 0], X[:, 1])
plt.scatter(X[X_outliers[0], 0], X[X_outliers[0], 1])
plt.xlabel("Latency")
plt.ylabel("Throughput")
plt.show()
```
